# Hands-on: Definição do tamanho da melhor janela para filtragem dos desvanecimentos do canal banda estreita pelo método do Vega

### Objetivos
As metas desse tutorial são ajudar o usuário a:
* Apresentar a geração de uma série temporal sintética com valores de potência recebida em função da distância, para caracterizar o Desvanecimento de Larga Escala (Perda de Percurso, Sombreamento) e Pequena Escala do canal;
* Compreender e aplicar a técnica do Vega utilizada para encontrar o tamanho da melhor janela de filtragem dos desvanecimentos de pequena e larga escala do canal.

## Prática 1: Geração do Canal Sintético

#### Passo 1: Criar Função para Geração de Canal Sintético

Crie um arquivo com nome fGeraCanal.m contendo o script apresentado na célula seguinte e realize as seguintes ações:
- Aalise e compreenda o funcionamento da função;
- Implemente a curva de distribuição Weibull (Plus).

 A função apresenta as seguintes entradas e saídas:
 
ENTRADAS:
   - fc: frequência da portadora
   - totalLength: distância final da rota de medição (m)
   - d0: distância de referência (m)
   - nPoints: número de amostras da rota de medição
   - dMed: distância entre pontos de medição (m)
   - P0: potência medida na distância de referência d0 (dBm)
   - n: expoente de perda de percurso
   - sigmaShad: desvio padrão do shadowing (dB)
   - nSampShadW: tamanho da janela de correlação do shadowing
   - txPower: potência de transmissão (dBm)
   - pdftype: tipo de pdf (Nakagami, Rician, Rayleigh, Weibull)
   - s1: parâmetro "sigma 1" da pdf Rician
   - s2: parâmetro "sigma 2" da pdf Rician
   - sigma: parâmetro "sigma" da pdf Rayleigh
   - m: parâmetro "m" da pdf Nakagami
   - a: parâmetro "a" da pdf Weibull
   - b: parâmetro "b" da pdf Weibull

SAÍDAS:
   - vtDist: distância dos pontos de medição (m)
   - vtPathLoss: amostras da perda de percurso (dBm)
   - vtShad: amostras do somrbeamento (dBm)
   - vtFastFading: amostras do desvanecimento de pequena escala (dBm)
   - vtPrxdBm: potência recebida com o canal completo (dBm)


```matlab
%%file fGeraCanal.m
function [vtDist, vtSlowFadMod, vtPathLossMod, vtShadMod, vtFastFadMod, vtPrx] = fGeraCanal(sPar)
% Parser dos parâmetros de entrada
fc = sPar.fc;                      % Frequência da Portadora  
totalLength = sPar.totalLength;    % Distância final da rota de medição (m)
d0 = sPar.d0;                      % distância de referência (m)
nPoints = sPar.nPoints;            % Número de amostras da rota de medição
dMed = (totalLength-d0)/(nPoints-1);   % Distância entre pontos de medição (m)
p0 = sPar.P0;                      % Potência medida na distância de referência d0 (dBm)
n = sPar.n;                        % Expoente de perda de percurso
sigmaShad = sPar.sigmaShad;        % Desvio padrão do shadowing em dB
nSampShadW = ceil(sPar.nSampShadW); % Tamanho da janela de correlação do shadowing
txPower = sPar.txPower;            % Potência de transmissão (dBm)
pdftype = sPar.pdftype;            % Tipo de pdf (Nakagami, Rician, Rayleigh, Weibull)
s1 = sPar.p;                       % Parâmetro "sigma 1" da pdf Rician
s2 = s1;                           % Parâmetro "sigma 2" da pdf Rician
sigma = sPar.p;                    % Parâmetro "sigma" da pdf Rayleigh
m = sPar.p;                        % Parâmetro "m" da pdf Nakagami
a = sPar.p;                        % Parâmetro "a" da pdf Weibull
b = sPar.b;                        % Parâmetro "b" da pdf Weibull

% Vetor de distância das medições em relação ao transmissor
vtDist = d0:dMed:totalLength;
nSamples = length(vtDist);

% Geração da Perda de percurso (determinística)
% Modelo COST 231 Considerando visada direta (LOS)
vtPathLossMod = p0 + (42.6 + 26.*log10(vtDist./d0) + 20.*log10(fc)); 

% Geração do Sombreamento
nShadowSamples = floor(nSamples/nSampShadW)+1;
shadowing = sigmaShad*randn(1,nShadowSamples);
% Amostras para a última janela
restShadowing = sigmaShad*randn(1,1)*ones(1,mod(nSamples,nSampShadW));
% Repetição do mesmo valor de sombreamento durante a janela de correlação
shadowing = ones(nSampShadW,1)*shadowing;
% Amostras organizadas em um vetor
shadowing = [reshape(shadowing,1,nShadowSamples*nSampShadW),restShadowing];
% Filtragem para evitar variação abrupta do sombreamento
jan = nSampShadW/2;
iCont = 1;
for i = jan+1:nSamples+nSampShadW-jan
    vtShadMod(iCont) = mean(shadowing(i-jan:i+jan)); %diminuir a variação brusca do sombreamento
    iCont = iCont+1;
end
% Ajuste do desvio padrão depois do filtro de correlação do sombreamento
vtShadMod = vtShadMod*std(shadowing)/std(vtShadMod);
vtShadMod = vtShadMod - mean(vtShadMod)+ mean(shadowing);

% Geração do desvanecimento de pequena escala
switch pdftype
    case 'Nakagami'
        pdfEquation = @(x)((2.*m.^m)./(gamma(m))).*x.^(2.*m-1).*exp(-(m.*x.^2));
    case 'Rayleigh'
        pdfEquation = @(x) (x./sigma.^2).*exp((-x.^2)/(2.*(sigma.^2)));
    case 'Weibull'
       % pdfEquation = ;
    case 'Rician'
        pdfEquation = @(x) (x ./ (s1.^2)) .*exp(-0.5 * (x.^2 + s2.^2) ./ ...
                      (s1.^2)) .*besseli(0, x .* s2 ./ (s1.^2));
end
normEnvelope = slicesample(1,nSamples,'pdf',pdfEquation);
vtFastFadMod=20.*log10(normEnvelope');

% Cálculo da Potência recebida
txPower = txPower*ones(1,nSamples);
vtPrx = txPower-vtPathLossMod+vtShadMod+vtFastFadMod; % Potência recebida
vtSlowFadMod = txPower-vtPathLossMod+vtShadMod; % Desvanecimento de Larga Escala

% Salva variáveis do canal no Matlab
save([sPar.chFileName '.mat'],'vtDist', 'vtSlowFadMod', 'vtPathLossMod', 'vtShadMod', 'vtFastFadMod', 'vtPrx');
```

### Passo 1: Cálculo da Curva $1\sigma$-Spread e Outros Parâmetros

Realize as seguintes ações:

**Passo 1.1 - Cálculo da Curva $1\sigma$-Spread:**
- Defina o vetor de tamanhos das janelas que serão testadas e inicie um laço for;
- Obtenha o vetor de médias estimadas (r), utilizando o filtro de janela móvel sobre o vetor de potência recebida do canal;
- Obtenha o vetor de médias verdadeiras aproximadas (m'), utilizando o filtro de janela móvel sobre o vetor de médias estimadas;  
- Obtenha o vetor das médias estimadas normalizadas, subtraindo o vetor de médias verdadeiras aproximadas do vetor de médias estimadas;
- Obtenha a curva de 1-sigma spread, efetuando o seguinte cálculo sobre o vetor das médias estimadas normalizadas: $1\sigma_{\hat m}Spread = 20*log\frac{1+\sigma_{\hat m_{norm}}}{1-\sigma_{\hat m_{norm}}} (dB) $, em que $\sigma_{\hat m_{norm}}$ é o desvio padrão do vetor das médias estimadas normalizadas.

**Passo 1.2 - Cálculo da Menor distância de Correlação entre as Amostras ("d"):**
- Faça a autocorrelação do vetor de médias estimadas normalizadas;
- Para cada tamanho de janela testada, obtenha os valores de "d" correspondente à correlação de 0.2 e ao primeiro nulo da curva de autocorrelação.
Obs.: Esses dados serão utilizados no Passo 2.

**Passo 1.3 - Estimação dos Desvanecimentos de Larga e Pequena Escala e Cálculos de RMSE:****
- Realize a estimação dos desvanecimentos de Larga e Pequena Escala para cada tamanho de janela;
- Calcule a raiz do erro médio quadrático (RMSE) dos desvanecimentos de Larga e Pequena Escala em relação ao canal modelo para cada tamanho de janela.

**Passo 1.4 - Plotes:**
- Escolha um valor de tamanho de janela e plote num mesmo gráfico as curvas de potência recebida do canal modelo, de potência da perda de percurso estimada e de potência da perda de percurso + sombreamento estimados;
- Plote a curva $1\sigma$-Spread;
- Plote as curvas de RMSE dos Desvanecimentos de Larga e de Pequena Escala em relação ao canal modelo para cada tamanho de janela;
- Plote das curvas de potência dos Desvanecimentos de Larga e de Pequena Escala para o menor RMSE.

**Passo 1.5 - Definição do Valor da Melhor Janela:**
- Verifique o melhor tamanho da janela no ponto de menor valor da curva 1-sigma spread.


```matlab
%----------------------------------------Passo 1.1-----------------------------------------------------
% Tamanhos das Janelas de estimação
vtW = [2:1:150]; 

for iw = 1:length(vtW) 
%% Cálculo das médias estimadas (r) [dBm]
vtPrx_mW = 10.^(vtPrx/10); % Transforma a potência de dBm em mWatts
vtEstMean = []; % Vetor para as médias estimadas (r)

% Cálculo da Média por Janela Móvel
dMeiaJanela = floor(vtW(iw)/2);  % Meia janela
nSamples = length(vtPrx_mW);
ij = 1;
for ik = dMeiaJanela + 1 : nSamples - dMeiaJanela
    vtEstMean(ij) = 10*log10(mean(vtPrx_mW(ik-dMeiaJanela:ik+dMeiaJanela)));
    ij = ij + 1;
end

%% Cálculo da Média Verdadeira Aproximada (m')
vtEstMean_mW = 10.^(vtEstMean/10); % Transforma a potência de dBm em mWatts
vtTrueMean = []; % vetor para a Média Verdadeira Aproximada

% Cálculo da média verdadeira aproximada utilizando Janela Móvel
lenEstMean=length(vtEstMean);
ij = 1; 
for ik = dMeiaJanela + 1 : lenEstMean - dMeiaJanela
    vtTrueMean(ij) = 10*log10(mean(vtEstMean_mW(ik-dMeiaJanela:ik+dMeiaJanela)));
    ij = ij + 1;
end
% Ajuste no tamanho dos vetores devido a filtragem
vtEstMeanAjust = vtEstMean( dMeiaJanela+1 : lenEstMean-dMeiaJanela );

% Cálculo das Médias Estimadas Normalizadas (r_norm) [dBm]
vtEstMeanNorm = vtEstMeanAjust - vtTrueMean;

% Cálculo do 1 sigma_m Spread
stdEstMeanNorm = std(10.^(vtEstMeanNorm/10));
sigmaSpread(iw) = 20*log10((1+stdEstMeanNorm)/(1-stdEstMeanNorm));

%----------------------------------------Passo 1.2-----------------------------------------------------
%% Autocorrelação para encontrar "d"
[aCorr, lags] = autocorr(vtEstMeanNorm,'NumLags',sPar.nPoints/100);

% Valor de d para uma correlação de 0.2
indexD0p2 = find(aCorr<=0.2, 1, 'first');
vtD0p2(iw) = (indexD0p2-1)*dMod; %vetor com os valores de "d" para cada tamanho de janela

% Valor de d para o first null
indexD0p0 = find(aCorr<=0, 1, 'first');
vtD0p0(iw) = (indexD0p0-1)*dMod; %vetor com os valores de "d" para cada tamanho de janela

%----------------------------------------Passo 1.3-----------------------------------------------------
% Estimação dos Desvanecimentos de Larga e Pequena Escala para cada tamanho de janela
%% Cálculo do desvanecimento lento (Larga Escala) [dBm]
% Desvanecimento de larga escala: perda de percurso + sombreamento 
vtSlowFading = []; % Vetor para o Desvanecimento de Larga Escala

% Média por Janela Móvel (Filtragem)
dMeiaJanela = floor(vtW(iw)/2);  % Meia janela
ij = 1;
for ik = dMeiaJanela + 1 : nSamples - dMeiaJanela
    vtSlowFading(ij) = 10*log10(mean(vtPrx_mW(ik-dMeiaJanela:ik+dMeiaJanela)));
    ij = ij + 1;
end

% Ajuste no tamanho dos vetores devido a filtragem
vtDistAjust = vtDist( dMeiaJanela+1 : nSamples-dMeiaJanela );
vtPrxAjust = vtPrx( dMeiaJanela+1 : nSamples-dMeiaJanela );
vtShadModAjust = vtShadMod(dMeiaJanela+1 : nSamples-dMeiaJanela );
vtSlowFadModAjust = vtSlowFadMod(dMeiaJanela+1 : nSamples-dMeiaJanela );
vtFastFadModAjust = vtFastFadMod(dMeiaJanela+1 : nSamples-dMeiaJanela );

%% Cálculo do Desvanecimento Rápido (Pequena Escala) [dBm]
vtFastFading = vtPrxAjust - vtSlowFading;

%% Cálculo da reta e coeficiente de perda de percurso
vtDistAjustLog = log10(vtDistAjust);
% Cálculo do coeficiente da reta que melhor caracteriza a perda de percurso
dCoefReta = polyfit(vtDistAjustLog,vtPrxAjust,1);
% Expoente de perda de percurso estimado
dNEst(iw) = -dCoefReta(1)/10;
% Perda de percurso estimada para os pontos de medição
vtPathLoss = polyval(dCoefReta,vtDistAjustLog);  

%% Cálculo do Desvanecimento por Sombreamento
vtShadowing = vtSlowFading - vtPathLoss;

%% Cálculos de RMSE
%Calcula erro do shadowing
rmseShad(iw) = sqrt( sum( (vtShadModAjust-vtShadowing).^2) / numel(vtShadModAjust) );
% Offset da média gerado pelo filtro de janela móvel
movAvOffset = mean(vtSlowFading)-mean(vtSlowFadModAjust); 
%Calcula erro do Slow Fading
rmseSlowFad(iw) = sqrt( sum( (vtSlowFadModAjust-vtSlowFading+movAvOffset).^2) / numel(vtSlowFadModAjust) );
%Calcula erro do Fast Fading
rmseFastFad(iw) = sqrt( sum( (vtFastFadModAjust-vtFastFading-movAvOffset).^2) / numel(vtFastFadModAjust) );

%----------------------------------------Passo 1.4-----------------------------------------------------
% Para plotar posteriormente o vetor de Desvanecimento de Pequena Escala 
vtSlowFadingPlot(iw,1:length(vtSlowFading)) = vtSlowFading;
vtSlowModAjustPlot(iw,1:length(vtSlowFading)) = vtSlowFadModAjust;
vtFastFadingPlot(iw,1:length(vtFastFading)) = vtFastFading;
vtFastModAjustPlot(iw,1:length(vtFastFading)) = vtFastFadModAjust;
vtShadowingPlot(iw,1:length(vtShadowing)) = vtShadowing;
vtShadModAjustPlot(iw,1:length(vtShadowing)) = vtShadModAjust;

%% Plot de Potência do Canal Real / Desvanecimentos Estimados 
plotOn=1; 
windowPlot = 30; % escolha o tamanho da janela para o plot
if plotOn==1
    if iw == windowPlot-1  
    figure;
    % Potência recebida do canal real
    plot(vtDistAjustLog,vtPrxAjust); hold all;
    % Potência recebida da perda de percurso
    plot(vtDistAjustLog,vtPathLoss,'linewidth', 2)
    % Potência recebida da perda de percurso e sombreamento
    plot(vtDistAjustLog,vtSlowFading,'linewidth', 2)
    xlabel('log_{10}(d)');
    ylabel('Potência [dBm]');
    legend('Prx Canal Modelo', 'Prx (somente perda de percurso)', 'Prx (perda de percurso + sombreamento)');
    title(strcat("Potência Recebida do Canal Estimado (W = ", num2str(vtW(iw)), ")"));
    end
end
end
```

### Passo 2: Cálculo da menor distância entre amostras descorrelacionadas ("d") para a Melhor Janela

Para calcular a menor distância entre amostras descorrelacionadas realize as seguintes ações:
- Obtenha o vetor de médias estimadas (r), utilizando o filtro de janela móvel, com o tamanho da melhor janela encontrada, sobre o vetor de potência recebida do canal;
- Obtenha o vetor de médias verdadeiras aproximadas (m'), utilizando o filtro de janela móvel com o tamanho da melhor janela encontrada, sobre o vetor de médias estimadas;  
- Obtenha o vetor das médias estimadas normalizadas, subtraindo o vetor de médias verdadeiras aproximadas do vetor de médias estimadas;
- Calcule a função de autocorrelação do vetor de médias estimadas normalizadas;
- A partir da função de autocorrelação, obtenha os valores de "d" para uma correlação de 0.2 e para o primeiro nulo;
- A partir dos dados calculados no Passo 1.2, plote o gráfico de "d" por "2L";
- Verificar no gráfico "d" x "2L" em qual faixa de 2L o "d" permanece com o mesmo valor aproximadamente;

```matlab
%% Distance d between uncorrelated samples
% Cálculo das médias estimadas (r) [dBm]
vtEstMeanBestW = []; % Vetor para as médias estimadas (r)

% Média por Janela Móvel
dMeiaJanela = floor(bestWindow/2);  % Meia janela
nSamples = length(vtPrx_mW);
ij = 1;
for ik = dMeiaJanela + 1 : nSamples - dMeiaJanela
    vtEstMeanBestW(ij) = 10*log10(mean(vtPrx_mW(ik-dMeiaJanela:ik+dMeiaJanela)));
    ij = ij + 1;
end

% Cálculo das Médias Verdadeiras Aproximadas (m')
vtEstMean_mW = 10.^(vtEstMeanBestW/10); % Transforma a potência de dBm em mWatts
vtTrueMeanBestW = []; % vetor para a Média Verdadeira Aproximada

% Janela Móvel para obter a média verdadeira aproximada
lenEstMean=length(vtEstMeanBestW);
ij = 1; 
for ik = dMeiaJanela + 1 : lenEstMean - dMeiaJanela
    vtTrueMeanBestW(ij) = 10*log10(mean(vtEstMean_mW(ik-dMeiaJanela:ik+dMeiaJanela)));
    ij = ij + 1;
end
% Ajuste no tamanho dos vetores devido a filtragem
vtEstMeanAjustBestW = vtEstMeanBestW( dMeiaJanela+1 : lenEstMean-dMeiaJanela );
% Cálculo das Médias Estimadas Normalizadas (r_norm) [dBm]
vtEstMeanNormBestW = vtEstMeanAjustBestW - vtTrueMeanBestW;
% Autocorrelação para encontrar "d"
[aCorrBestW, lagsBW] = autocorr(10.^(vtEstMeanNormBestW/10),'NumLags',sPar.nPoints/100);

% Valor de d para uma correlação de 0.2
indexD0p2 = find(aCorrBestW<=0.2, 1, 'first');
d0p2 = (indexD0p2-1)*dMod;
disp(strcat("A distância entre as amostras não correl. (d) (correl. 0.2) é: ", num2str(d0p2/lambda)," ", char(955)))

% Valor de d para o first null
indexD0p0 = find(aCorrBestW<=0, 1, 'first');
d0p0 = (indexD0p0-1)*dMod; %vetor com os valores de "d" para cada tamanho de janela
disp(strcat("A distância entre as amostras não correl. (d) (first null)  é: ", num2str(d0p0/lambda)," ", char(955)))
```

```matlab
% Plot da Função de Correlação
figure;
plot(lagsBW*dMod/lambda,aCorrBestW); hold on
plot((indexD0p2-1)*dMod/lambda,aCorrBestW(indexD0p2), 'o', 'linewidth', 1, 'Color',[0.9 0.5 0.9])
plot((indexD0p0-1)*dMod/lambda,aCorrBestW(indexD0p0), 'o', 'linewidth', 1, 'Color',[0.0 0.7 0.1])
hold off
pt1 = sprintfc('(%1.2f, %1.2f)', [(d0p2/lambda); aCorrBestW(indexD0p2)].');
pt2 = sprintfc('(%1.2f, %1.2f)', [(d0p0/lambda); aCorrBestW(indexD0p0)].');
text((d0p2/lambda), aCorrBestW(indexD0p2), pt1,'FontSize',8,...
    'color', [0.9 0.5 0.9], 'HorizontalAlignment','left', 'VerticalAlignment','top')
text((d0p0/lambda), aCorrBestW(indexD0p0), pt2,'FontSize',8,...
    'color', [0.0 0.7 0.1], 'HorizontalAlignment','left', 'VerticalAlignment','top')
ylabel('Correlação');
xlabel('d (\lambda)');
legend('Autocorrelation', '0.2 Correlation', 'First Null');
title("Autocorrelação da Média Estimada Normalizada para a Melhor Janela")
```

```matlab
% Plot dos valores de "d" x "2L"
figure; hold on
plot((vtW-1)*dMod/lambda,vtD0p2/lambda, 'linewidth', 2, 'Color',[0.9 0.5 0.9])
plot((vtW-1)*dMod/lambda,vtD0p0/lambda, 'linewidth', 2, 'Color',[0.0 0.7 0.1])
y1=get(gca,'ylim');
plot([(vtW(minIndxSigma)-1)*dMod/lambda,(vtW(minIndxSigma)-1)*dMod/lambda],[y1(1), y1(2)], 'linewidth', 1, 'Color',[0.9 0.1 0.0])
plot([(vtW(minIndxSlow)-1)*dMod/lambda,(vtW(minIndxSlow)-1)*dMod/lambda],[y1(1), y1(2)], 'linewidth', 1, 'Color',[0.9 0.5 0.0])
plot([(vtW(minIndxFast)-1)*dMod/lambda,(vtW(minIndxFast)-1)*dMod/lambda],[y1(1), y1(2)], 'linewidth', 1, 'Color',[0.5 0.5 0.0])
hold off
ylabel('d (\lambda)');
xlabel('Window Size - 2L (\lambda)');
legend('0.2 Correlation', 'First Null', 'min(1 \sigma spread)', 'min(Shadowing RMSE)', 'min(Fast Fading RMSE)');
title("d x 2L")
```

Observa-se que o valor de "d" apresenta a menor variação até o tamanho da melhor janela (para a correlação no first null) e até o tamanho da janela relativa ao mínimo RMSE do desvanecimento rápido (para a correlação de 0.2).

### Passo 3: Cálculo do número mínimo de amostras na janela móvel ("N")

Realize as seguintes ações:
- Calcule o valor de "N" para 0.2 de correlação e para a correlação no first null, utilizando a seguinte fórmula: $N = \frac{2L}{d}$.
- Calcule o valor de "N" por meido da fórmula utilizada pelo Parsons e adotada pelo Vega. Considerando um receptor com característica linear, um máximo erro de + ou - 1dB em torno da média e 90% de grau de confiança, Parsons obteve $N$ por meio da formulação a seguir.
- Realizar o cálculo de $N$ para diversas rotas e selecionar o mais frequênte. Para isso, implementar um laço for com a geração de várias realizações do canal sintético (**Plus**). 

$20 \log(m+1.65 \frac{\sigma_{r}}{\sqrt{N}}) - 20 \log(m-1.65 \frac{\sigma_{r}}{\sqrt{N}}) \leq 2dB$

Simplificando, temos:

$\frac{1.65 (10^{0.1}+1) \sigma_{r}}{(10^{0.1}-1) m} \leq \sqrt{N}$

Que resulta em:

$207,22 (\frac{\sigma_{r}}{m})^{2} \leq N$

O número mínimo de amostras necessário para realizar as estimações das médias ($N$) depende do desvio padrão ($\sigma_{r}$), da média verdadeira dos dados ($m$) e do grau de confiança.


```matlab
% Necessary number N of measuring points
% Cálculo do valor de "N" para 0.2 de correlação
n0 = bestWindow/indexD0p0;
disp(strcat("A quantidade mínima de pontos de medição  (N) (first null)  é: ", num2str(n0)))

% Cálculo do valor de "N" para 0.2 de correlação
n2 = bestWindow/indexD0p2;
disp(strcat("A quantidade mínima de pontos de medição  (N) (correl 0.2)  é: ", num2str(n2)))

% Cálculo média verdadeira
m = mean(10.^(vtTrueMean/10));

% Cálculo do desvio padrão
s = std(10.^(vtPrx/10));

% Cálculo de "N" segundo Parsons
N = 207.22*((s/m)^2);

disp(strcat("A quantidade mínima de pontos de medição  (N) (Parsons/VEGA)é: ", num2str(N)))
```
