# Hands-on: Definição do tamanho da melhor janela para filtragem dos desvanecimentos do canal banda estreita (Método do Vega Aprimorado)

### Objetivos
As metas desse tutorial são ajudar o usuário a:
* Compreender e aplicar a técnica do Vega utilizada para encontrar o tamanho da melhor janela de filtragem dos desvanecimentos de pequena e larga escala do canal;
* Apresentar uma possibilidade de novo método para definir a melhor janela para filtragem dos desvanecimentos do canal.

## Prática 1: Geração do Canal Sintético

#### Passo 1: Criar Função para Geração de Canal Sintético

Crie um arquivo com nome fGeraCanal.m contendo o script apresentado na célula seguinte e realize as seguintes ações:
- Aalise e compreenda o funcionamento da função;
- Implemente a curva de distribuição Weibull (Plus).

 A função apresenta as seguintes entradas e saídas:
 
ENTRADAS:
   - fc: frequência da portadora
   - totalLength: distância final da rota de medição (m)
   - d0: distância de referência (m)
   - nPoints: número de amostras da rota de medição
   - dMed: distância entre pontos de medição (m)
   - P0: potência medida na distância de referência d0 (dBm)
   - n: expoente de perda de percurso
   - sigmaShad: desvio padrão do shadowing (dB)
   - nSampShadW: tamanho da janela de correlação do shadowing
   - txPower: potência de transmissão (dBm)
   - pdftype: tipo de pdf (Nakagami, Rician, Rayleigh, Weibull)
   - s1: parâmetro "sigma 1" da pdf Rician
   - s2: parâmetro "sigma 2" da pdf Rician
   - sigma: parâmetro "sigma" da pdf Rayleigh
   - m: parâmetro "m" da pdf Nakagami
   - a: parâmetro "a" da pdf Weibull
   - b: parâmetro "b" da pdf Weibull

SAÍDAS:
   - vtDist: distância dos pontos de medição (m)
   - vtPathLoss: amostras da perda de percurso (dBm)
   - vtShad: amostras do somrbeamento (dBm)
   - vtShortFading: amostras do desvanecimento de pequena escala (dBm)
   - vtPrxdBm: potência recebida com o canal completo (dBm)


```matlab
%%file fGeraCanal.m
function [vtDist, vtLongFadMod, vtPathLossMod, vtShadMod, vtShortFadMod, vtPrx] = fGeraCanal(sPar)
% Parser dos parâmetros de entrada
fc = sPar.fc;                      % Frequência da Portadora  
totalLength = sPar.totalLength;    % Distância final da rota de medição (m)
d0 = sPar.d0;                      % distância de referência (m)
nPoints = sPar.nPoints;            % Número de amostras da rota de medição
dMed = (totalLength-d0)/(nPoints-1);   % Distância entre pontos de medição (m)
p0 = sPar.P0;                      % Potência medida na distância de referência d0 (dBm)
n = sPar.n;                        % Expoente de perda de percurso
sigmaShad = sPar.sigmaShad;        % Desvio padrão do shadowing em dB
nSampShadW = ceil(sPar.nSampShadW); % Tamanho da janela de correlação do shadowing
txPower = sPar.txPower;            % Potência de transmissão (dBm)
pdftype = sPar.pdftype;            % Tipo de pdf (Nakagami, Rician, Rayleigh, Weibull)
s1 = sPar.p;                       % Parâmetro "sigma 1" da pdf Rician
s2 = s1;                           % Parâmetro "sigma 2" da pdf Rician
sigma = sPar.p;                    % Parâmetro "sigma" da pdf Rayleigh
m = sPar.p;                        % Parâmetro "m" da pdf Nakagami
a = sPar.p;                        % Parâmetro "a" da pdf Weibull
b = sPar.b;                        % Parâmetro "b" da pdf Weibull

% Vetor de distância das medições em relação ao transmissor
vtDist = d0:dMed:totalLength;
nSamples = length(vtDist);

% Geração da Perda de percurso (determinística)
% Modelo COST 231 Considerando visada direta (LOS)
vtPathLossMod = p0 + (42.6 + 26.*log10(vtDist./d0) + 20.*log10(fc)); 

% Geração do Sombreamento
nShadowSamples = floor(nSamples/nSampShadW)+1;
shadowing = sigmaShad*randn(1,nShadowSamples);
% Amostras para a última janela
restShadowing = sigmaShad*randn(1,1)*ones(1,mod(nSamples,nSampShadW));
% Repetição do mesmo valor de sombreamento durante a janela de correlação
shadowing = ones(nSampShadW,1)*shadowing;
% Amostras organizadas em um vetor
shadowing = [reshape(shadowing,1,nShadowSamples*nSampShadW),restShadowing];
% Filtragem para evitar variação abrupta do sombreamento
jan = nSampShadW/2;
iCont = 1;
for i = jan+1:nSamples+nSampShadW-jan
    vtShadMod(iCont) = mean(shadowing(i-jan:i+jan)); %diminuir a variação brusca do sombreamento
    iCont = iCont+1;
end
% Ajuste do desvio padrão depois do filtro de correlação do sombreamento
vtShadMod = vtShadMod*std(shadowing)/std(vtShadMod);
vtShadMod = vtShadMod - mean(vtShadMod)+ mean(shadowing);

% Geração do desvanecimento de pequena escala
switch pdftype
    case 'Nakagami'
        pdfEquation = @(x)((2.*m.^m)./(gamma(m))).*x.^(2.*m-1).*exp(-(m.*x.^2));
    case 'Rayleigh'
        pdfEquation = @(x) (x./sigma.^2).*exp((-x.^2)/(2.*(sigma.^2)));
    case 'Weibull'
       % pdfEquation = ;
    case 'Rician'
        pdfEquation = @(x) (x ./ (s1.^2)) .*exp(-0.5 * (x.^2 + s2.^2) ./ ...
                      (s1.^2)) .*besseli(0, x .* s2 ./ (s1.^2));
end
normEnvelope = slicesample(1,nSamples,'pdf',pdfEquation);
vtShortFadMod=20.*log10(normEnvelope');

% Cálculo da Potência recebida
txPower = txPower*ones(1,nSamples);
vtPrx = txPower-vtPathLossMod+vtShadMod+vtShortFadMod; % Potência recebida
vtLongFadMod = txPower-vtPathLossMod+vtShadMod; % Desvanecimento de Larga Escala

% Salva variáveis do canal no Matlab
save([sPar.chFileName '.mat'],'vtDist', 'vtLongFadMod', 'vtPathLossMod', 'vtShadMod', 'vtShortFadMod', 'vtPrx');
```

#### Passo 2: Geração de Canais Sintéticos

Crie um arquivo com o nome "canaisPdfs.m" e realize as seguintes ações:
- Defina os parâmetros de entrada do canal;
- Chame a função "fGeraCanal.m";
- Plote num mesmo gráfico as curvas de potência recebida do canal modelo, de potência da perda de percurso e de potência da perda de percurso + sombreamento;
- Faça o plote acima para as distribuições Nakagami, Rician, Rayleigh, Weibull;
- Gere e plote as PDFs de cada distribuição num mesmo gráfico (**Plus**).

```matlab
% Parâmetros para geração do canal sintético
sPar.fc = 1.359;                          % Carrier frequency (MHz)
cc = 3e8;                                 % Light speed (m/s) 
lambda = cc/(sPar.fc*1e6);                % Wavelength (m)
dMod = 0.05*lambda;                       % Sampling spacing (m)
sPar.lenShadWindow = 10*lambda;           % Tamanho da janela de correlação do shadowing (colocar em função da distância de correlação)
sPar.nSampShadW = sPar.lenShadWindow/dMod;% Número de amostras numa janela de correlação do shadowing
sPar.d0 = 5;                              % distância de referência d0 (m)
sPar.P0 = 0;                              % Potência medida na distância de referência d0 (em dBm)
sPar.nPoints = 50000;                     % Número de amostras da rota de medição
sPar.totalLength = sPar.nPoints*dMod;     % Distância final da rota de medição (m)
sPar.n = 4;                               % Expoente de perda de percurso
sPar.sigmaShad = 6;                       % Desvio padrão do shadowing em dB
sPar.p = 4;                               % Parâmetro da pdf: sigma(Rayleigh); m (Nakagami); a(Weibull); s(Rician); ...)
sPar.b = 1;                               % Parâmetro "b" da pdf Weibull
sPar.txPower = 0;                         % Potência de transmissão (dBm)
sPar.chFileName  = 'Prx_sintetic';        % Nome do arquivo gerado com as amostras do canal
sPar.pdftype = 'Rayleigh';                % Tipo de pdf (Nakagami, Rician, Rayleigh, Weibull)
%sPar.pdftype = 'Nakagami';
%sPar.pdftype = 'Rician';
```

```matlab
% Chama função que gera o canal sintético
[vtDist, vtLongFadMod, vtPathLossMod, vtShadMod, vtShortFadMod, vtPrx] = fGeraCanal(sPar);

% Passa o vetor de distância para a escala logarítmica
vtDistLog = log10(vtDist);

% Calcula o desvanecimento de larga escala (perda de percurso + sombreamento)
vtLongFadingMod = sPar.txPower-vtPathLossMod+vtShadMod;

% Offset para alinhar os desvanecimentos de larga escala
shortFadMean = mean(vtShortFadMod);

figure;
% Potência recebida do canal real
plot(vtDistLog,vtPrx); hold all;
% Potência recebida da perda de percurso
plot(vtDistLog,sPar.txPower-vtPathLossMod+shortFadMean,'linewidth', 2)
% Potência recebida da perda de percurso + sombreamento
plot(vtDistLog,vtLongFadingMod+shortFadMean,'linewidth', 2)
xlabel('log_{10}(d)');
ylabel('Potência [dBm]');
legend('Prx Canal Modelo', 'Prx (somente perda de percurso)', 'Prx (perda de percurso + sombreamento)');
title(strcat("Potência Recebida do Canal Modelo"));
```

```matlab
% Parâmetros para geração do canal sintético
sPar.fc = 1.359;                          % Carrier frequency (MHz)
cc = 3e8;                                 % Light speed (m/s) 
lambda = cc/(sPar.fc*1e6);                % Wavelength (m)
dMod = 0.05*lambda;                       % Sampling spacing (m)
sPar.lenShadWindow = 10*lambda;           % Tamanho da janela de correlação do shadowing (colocar em função da distância de correlação)
sPar.nSampShadW = sPar.lenShadWindow/dMod;% Número de amostras numa janela de correlação do shadowing
sPar.d0 = 5;                              % distância de referência d0 (m)
sPar.P0 = 0;                              % Potência medida na distância de referência d0 (em dBm)
sPar.nPoints = 50000;                     % Número de amostras da rota de medição
sPar.totalLength = sPar.nPoints*dMod;     % Distância final da rota de medição (m)
sPar.n = 4;                               % Expoente de perda de percurso
sPar.sigmaShad = 6;                       % Desvio padrão do shadowing em dB
sPar.p = 4;                               % Parâmetro da pdf: sigma(Rayleigh); m (Nakagami); a(Weibull); s(Rician); ...)
sPar.b = 1;                               % Parâmetro "b" da pdf Weibull
sPar.txPower = 0;                         % Potência de transmissão (dBm)
sPar.chFileName  = 'Prx_sintetic';        % Nome do arquivo gerado com as amostras do canal
sPar.pdftype = 'Rayleigh';                % Tipo de pdf (Nakagami, Rician, Rayleigh, Weibull)
%sPar.pdftype = 'Nakagami';
%sPar.pdftype = 'Rician';
%sPar.pdftype = 'Weibull';
```

```matlab
% Chama função que gera o canal sintético
[vtDist, vtLongFadMod, vtPathLossMod, vtShadMod, vtShortFadMod, vtPrx] = fGeraCanal(sPar);
```

### Passo 1: Cálculo da Curva $1\sigma$-Spread para os Desvanecimentos de Larga e Pequena Escala

Realize as seguintes ações:

**Passo 1.1 - Estimação dos Desvanecimentos de Larga e Pequena Escala e Cálculos de RMSE:**
- Defina o vetor de tamanhos das janelas que serão testadas e inicie um laço for;
- Realize a estimação dos desvanecimentos de Larga e Pequena Escala para cada tamanho de janela;
- Calcule a raiz do erro médio quadrático (RMSE) dos desvanecimentos de Larga e Pequena Escala em relação ao canal modelo para cada tamanho de janela.

**Passo 1.2 - Cálculo da curva $1\sigma$-Spread para o Desvanecimento de Larga Escala:**
- Obtenha o vetor de médias estimadas (r), observe que é o mesmo vetor de desvanecimento de larga escala já calculado no passo anterior;
- Obtenha o vetor de médias verdadeiras aproximadas (m'), utilizando o filtro de janela móvel sobre o vetor de médias estimadas; 
- Obtenha o vetor das médias estimadas normalizadas, subtraindo o vetor de médias verdadeiras aproximadas do vetor de médias estimadas;
- Obtenha a curva de $1\sigma$-Spread, efetuando o seguinte cálculo sobre o vetor das médias estimadas normalizadas: $1\sigma_{\hat m}Spread = 20*log\frac{1+\sigma_{\hat m_{norm}}}{1-\sigma_{\hat m_{norm}}} (dB) $, em que $\sigma_{\hat m_{norm}}$ é o desvio padrão do vetor das médias estimadas normalizadas.

**Passo 1.3 - Cálculo da curva $1\sigma$-Spread para o Desvanecimento de Pequena Escala:**
- Obtenha a curva de $1\sigma$-Spread, efetuando o seguinte cálculo sobre o vetor de desvanecimento de pequena escala: $1\sigma_{r_{0}}Spread = 20*log\frac{1+\sigma_{r_{0}}}{1-\sigma_{r_{0}}} (dB) $, em que $\sigma_{r_{0}}$ é o desvio padrão do vetor de desvanecimento de pequena escala.

**Passo 1.4 - Plotes:**
- Escolha um valor de tamanho de janela e plote num mesmo gráfico as curvas de potência recebida do canal modelo, de potência da perda de percurso estimada e de potência da perda de percurso + sombreamento estimados;
- Plote as curvas de RMSE do sombreamento e de Pequena Escala em relação ao canal modelo para cada tamanho de janela;

```matlab
%----------------------------------------Passo 1.1-----------------------------------------------------
% Estimação dos Desvanecimentos de Larga e Pequena Escala para cada tamanho de janela
% Tamanhos das Janelas de estimação
vtW = [2:1:150]; 

%% Cálculo do Desvanecimento de Larga Escala = perda de percurso + sombreamento 
% O Desvanecimento de larga escala é denominado por Vega de Médias Estimadas
for iw = 1:length(vtW) 
vtPrx_mW = 10.^(vtPrx/10); % Transforma a potência de dBm em mWatts
vtLongFading = []; % Vetor para Desvanecimento de larga escala 

% Cálculo das médias estimadas (r) por Janela Móvel [dBm]
dMeiaJanela = floor(vtW(iw)/2);  % Meia janela
nSamples = length(vtPrx_mW);
ij = 1;
for ik = dMeiaJanela + 1 : nSamples - dMeiaJanela
    vtLongFading(ij) = 10*log10(mean(vtPrx_mW(ik-dMeiaJanela:ik+dMeiaJanela)));
    ij = ij + 1;
end

% Ajuste no tamanho dos vetores devido a filtragem
vtDistAjust = vtDist( dMeiaJanela+1 : nSamples-dMeiaJanela );
vtPrxAjust = vtPrx( dMeiaJanela+1 : nSamples-dMeiaJanela );
vtShadModAjust = vtShadMod(dMeiaJanela+1 : nSamples-dMeiaJanela );
vtLongFadModAjust = vtLongFadMod(dMeiaJanela+1 : nSamples-dMeiaJanela );
vtShortFadModAjust = vtShortFadMod(dMeiaJanela+1 : nSamples-dMeiaJanela );

%% Cálculo do Desvanecimento Pequena Escala [dBm]
vtShortFading = vtPrxAjust - vtLongFading;

%% Cálculo da reta e coeficiente de perda de percurso
vtDistAjustLog = log10(vtDistAjust);
% Cálculo do coeficiente da reta que melhor caracteriza a perda de percurso
dCoefReta = polyfit(vtDistAjustLog,vtPrxAjust,1);
% Expoente de perda de percurso estimado
dNEst(iw) = -dCoefReta(1)/10;
% Perda de percurso estimada para os pontos de medição
vtPathLoss = polyval(dCoefReta,vtDistAjustLog);  

%% Cálculo do Desvanecimento por Sombreamento
vtShadowing = vtLongFading - vtPathLoss;

%% Cálculos de RMSE
%Calcula erro do shadowing
rmseShad(iw) = sqrt( sum( (vtShadModAjust-vtShadowing).^2) / numel(vtShadModAjust) );
% Offset da média gerado pelo filtro de janela móvel
movAvOffset = mean(vtLongFading)-mean(vtLongFadModAjust); 
%Calcula erro do Long-term Fading
rmseLongFad(iw) = sqrt( sum( (vtLongFadModAjust-vtLongFading+movAvOffset).^2) / numel(vtLongFadModAjust) );
%Calcula erro do Short Fading
rmseShortFad(iw) = sqrt( sum( (vtShortFadModAjust-vtShortFading-movAvOffset).^2) / numel(vtShortFadModAjust) );


%----------------------------------------Passo 1.2-----------------------------------------------------
%% Cálculo da Média Verdadeira Aproximada (m')
vtEstMean = vtLongFading; % são vetores iguais
vtEstMean_mW = 10.^(vtEstMean/10); % Transforma a potência de dBm em mWatts
vtTrueMean = []; % vetor para a Média Verdadeira Aproximada

% Cálculo da média verdadeira aproximada utilizando Janela Móvel
lenEstMean=length(vtEstMean);
ij = 1; 
for ik = dMeiaJanela + 1 : lenEstMean - dMeiaJanela
    vtTrueMean(ij) = 10*log10(mean(vtEstMean_mW(ik-dMeiaJanela:ik+dMeiaJanela)));
    ij = ij + 1;
end
% Ajuste no tamanho dos vetores devido a filtragem
vtEstMeanAjust = vtEstMean( dMeiaJanela+1 : lenEstMean-dMeiaJanela );

% Cálculo das Médias Estimadas Normalizadas (r_norm) [dBm]
vtEstMeanNorm = vtEstMeanAjust - vtTrueMean;

% Cálculo do 1 sigma_m Spread para o Desvanecimento de Larga Escala
stdEstMeanNorm = std(10.^(vtEstMeanNorm/10));  % desvio padrão das Médias Estimadas Normalizadas 
sigmaSpreadLong(iw) = 20*log10((1+stdEstMeanNorm)/(1-stdEstMeanNorm));

%----------------------------------------Passo 1.3-----------------------------------------------------
%Calcula 1-sigma spread do Desvanecimento de Pequena Escala
vtShortFading_mW = 10.^(vtShortFading/10); % passa para a escala linear
vtShortFadingNorm = vtShortFading - 10*log10(mean(vtShortFading_mW)); % normalização do desvanecimento
stdShortFad(iw) = std(10.^(vtShortFadingNorm/10)); % desvio padrão do Desvanecimento de Pequena Escala estimado
sigmaSpreadShort(iw) = real(20*log10((1+stdShortFad(iw))/(1-stdShortFad(iw)))); % 1-sigma spread


%----------------------------------------Passo 1.4-----------------------------------------------------
%% Plot de Potência do Canal Real / Desvanecimentos Estimados 
plotOn=1; 
windowPlot = 30; % escolha o tamanho da janela para o plot
if plotOn==1
    if iw == windowPlot-1  
    figure;
    % Potência recebida do canal real
    plot(vtDistAjustLog,vtPrxAjust); hold all;
    % Potência recebida da perda de percurso
    plot(vtDistAjustLog,vtPathLoss,'linewidth', 2)
    % Potência recebida da perda de percurso e sombreamento
    plot(vtDistAjustLog,vtLongFading,'linewidth', 2)
    xlabel('log_{10}(d)');
    ylabel('Potência [dBm]');
    legend('Prx Canal Modelo', 'Prx (somente perda de percurso)', 'Prx (perda de percurso + sombreamento)');
    title(strcat("Potência Recebida do Canal Estimado (W = ", num2str(vtW(iw)), ")"));
    end
end
end

% 1.4.1 - Curvas de potência recebida do canal modelo, de potência da perda de percurso estimada e 
% de potência da perda de percurso + sombreamento estimados;
```

#### 1.4.2 -  Plot das curvas de RMSE para o Sombreamento e Desvancimento de Pequena Escala

 - Aqui são plotadas as curvas com os valores de RMSE entre os desvanecimentos estimados após a filtragem com cada tamanho de janela e os desvanecimentos do modelo de canal gerado.

```matlab
% Plot das curvas de RMSE do sombreamento e de Pequena Escala em relação ao canal modelo para cada tamanho de janela;
% Menores valores das curvas de RMSE do sombreamento, Desvanecimento de Larga escala e de Pequena Escala
[minRmseShad, minIndxShad]=min(rmseShad);
[minRmseLong, minIndxLong]=min(rmseLongFad);
[minRmseShort, minIndxShort]=min(rmseShortFad);

% Plot da curva de RMSE do sombreamento
figure;
plot((vtW-1)*dMod/lambda,rmseShad,'linewidth', 1, 'Color',[0.7 0.4 0.0]); hold all;
plot((vtW(minIndxShad)-1)*dMod/lambda, rmseShad(minIndxShad), 'o', 'linewidth', 1, 'Color',[0.9 0.5 0.0])
pt1 = sprintfc('(%1.2f, %1.2f)', [(vtW(minIndxShad)-1)*dMod/lambda; rmseShad(minIndxShad)].');
text((vtW(minIndxShad)-1)*dMod/lambda, rmseShad(minIndxShad), pt1,'FontSize',8,...
    'color', [0.9 0.5 0.0], 'HorizontalAlignment','left', 'VerticalAlignment','top')
ylabel('RMSE of Shadowing');
xlabel('Window Size - 2L (\lambda)');
legend('RMSE of Shadowing', 'min(RMSE of Shadowing)');
title("RMSE of Shadowing")

%% Plot do RMSE do Desvanecimento de Larga Escala
figure;
plot((vtW-1)*dMod/lambda,rmseLongFad,'linewidth', 1, 'Color',[0.7 0.7 0.2]); hold all;
plot((vtW(minIndxLong)-1)*dMod/lambda, minRmseLong, 'o', 'linewidth', 1, 'Color',[0.5 0.5 0.0])
pt1 = sprintfc('(%1.2f, %1.2f)', [(vtW(minIndxLong)-1)*dMod/lambda; minRmseLong].');
text((vtW(minIndxLong)-1)*dMod/lambda, minRmseLong, pt1,'FontSize',8,...
    'color', [0.5 0.5 0.0], 'HorizontalAlignment','left', 'VerticalAlignment','top')
ylabel('RMSE of Shadowing');
xlabel('Window Size - 2L (\lambda)');
legend('RMSE of Long-term Fading', 'min(RMSE of Long-term Fading)');
title("RMSE of Long-term Fading")

%% Plot do RMSE do Desvanecimento de Pequena Escala
figure;
plot((vtW-1)*dMod/lambda,rmseShortFad,'linewidth', 1,'color',[0.0 0.8 0.5]); hold all;
plot((vtW(minIndxShort)-1)*dMod/lambda, rmseShortFad(minIndxShort), 'o', 'linewidth', 1, 'Color',[0.5 0.5 0.0])
pt1 = sprintfc('(%1.2f, %1.2f)', [(vtW(minIndxShort)-1)*dMod/lambda; rmseShortFad(minIndxShort)].');
text((vtW(minIndxShort)-1)*dMod/lambda, rmseShortFad(minIndxShort), pt1,'FontSize',8,...
    'color', [0.5 0.5 0.0], 'HorizontalAlignment','left', 'VerticalAlignment','top')
ylabel('RMSE of Short-term Fading');
xlabel('Window Size - 2L (\lambda)');
legend('RMSE of Short-term Fading', 'min(RMSE of Short-term Fading)');
title("RMSE of Short-term Fading")
```

**Análises:**
- Observa-se que há um melhor valor para o tamanho da janela que proporciona o menor erro de estimação dos desvanecimentos de larga escala e pequena escala, considerando o modelo de canal gerado como referência;
- A estimação do sombreamento apresenta um maior erro, pois envolve a imprecisão da definição da reta de perda de percuso, subtraída do desvanecimento de larga escala.

### Passo 2: Definição do Valor da Melhor Janela

- Verifique o melhor tamanho da janela para a curva $1\sigma$-Spread (Long-term fading) no ponto de menor valor; 
- Verifique o melhor tamanho da janela para a curva $1\sigma$-Spread (Short-term fading) no primeiro ponto de inflexão; 

  - Para obter o ponto de inflexão, calcule a segunda derivada da curva $1\sigma$-Spread (Short-term fading) e selecione o primeiro ponto onde ela é igual a zero;
  - Plot a curva $1\sigma$-spread (Short-term fading) sinalizando o primeiro ponto de inflexão;
- Plote a curva de $1\sigma$-Spread para o Desvanecimento de Larga Escala (Long-term fading);
- Plote a curva de $1\sigma$-Spread para o Desvanecimento de Pequena Escala (Short-term fading).

```matlab
%% Best Window 1-sigma Spread
% Encontra o tamanho ótimo da janela - Larga Escala (menor valor da curva de 1 sigma spread)
[minSigmaSpreadLong,minIndxSigmaLong] = min(sigmaSpreadLong);

% Encontra o tamanho ótimo da janela - Pequena Escala (ponto de Inflexão)
[maxSigmaSpreadShort,maxIndxSigmaShort] = max(sigmaSpreadShort);

% Downsampling para diminuir o serrilhado e evitar multiplicidade dos pontos de inflexão
nDown = 7; % taxa de downsample
x=downsample(vtW(1:maxIndxSigmaShort),nDown);
yD=downsample(sigmaSpreadShort(1:maxIndxSigmaShort),nDown); 
y=smooth(x,yD,nDown)'; % suavisa a curva
% Cálculo da Segunda Derivada para obter o ponto de inflexão
dydx = gradient(y) ./ gradient(x); % derivada
dydx2 = gradient(dydx) ./ gradient(x); % segunda derivada
zci = @(v) find(v(:).*circshift(v(:), [-1 0]) <= 0); % retorna posições aproximadas dos cruzamentos = 0
zxidx = zci(dydx2); % ídices dos pontos de inflexão
% Tamanho da melhor janela no primeiro ponto de inflexão
infIndxSigmaShort = nDown*zxidx(1); 
```

```matlab
%% Plot do Ponto de Inflexão
figure;
plot((vtW-1)*dMod/lambda, sigmaSpreadShort)
hold on
plot((x-1)*dMod/lambda, yD)
plot((x-1)*dMod/lambda, y)
plot((x(zxidx(1))-1)*dMod/lambda, y(zxidx(1)), 'pg', 'MarkerFaceColor','g')
hold off
ylabel('1\sigma Spreading (dB)');
xlabel('Window Size - 2L (\lambda)');
legend('1\sigma Spread - Short-term Fading', 'Downsampled', 'Smoothed', 'Inflexion Point', 'Location','southeast')
title('1\sigma Spread - Short-term Fading')
```

```matlab
%% Plot 1-sigma Spread - Long-term Fading
figure;
plot((vtW-1)*dMod/lambda,sigmaSpreadLong); hold all;
plot((vtW(minIndxSigmaLong)-1)*dMod/lambda, minSigmaSpreadLong, 'o', 'linewidth', 1, 'Color',[0.9 0.1 0.0])
plot((vtW(infIndxSigmaShort)-1)*dMod/lambda, sigmaSpreadLong(infIndxSigmaShort), 'o', 'linewidth', 1, 'Color',[0.9 0.1 0.8])
plot((vtW(minIndxShort)-1)*dMod/lambda, sigmaSpreadLong(minIndxShort), 'o', 'linewidth', 1, 'Color',[0.5 0.5 0.0])
hold off
pt1 = sprintfc('(%1.2f, %1.2f)', [(vtW(minIndxSigmaLong)-1)*dMod/lambda; minSigmaSpreadLong].');
pt2 = sprintfc('(%1.2f, %1.2f)', [(vtW(infIndxSigmaShort)-1)*dMod/lambda; sigmaSpreadLong(infIndxSigmaShort)].');
pt3 = sprintfc('(%1.2f, %1.2f)', [(vtW(minIndxShort)-1)*dMod/lambda; sigmaSpreadLong(minIndxShort)].');
text((vtW(minIndxSigmaLong)-1)*dMod/lambda, sigmaSpreadLong(minIndxSigmaLong), pt1,'FontSize',8,...
    'color', [0.9 0.1 0.0], 'HorizontalAlignment','left', 'VerticalAlignment','top')
text((vtW(infIndxSigmaShort)-1)*dMod/lambda, sigmaSpreadLong(infIndxSigmaShort), pt2,'FontSize',8,...
    'color', [0.9 0.1 0.8], 'HorizontalAlignment','right', 'VerticalAlignment','bottom')
text((vtW(minIndxShort)-1)*dMod/lambda, sigmaSpreadLong(minIndxShort), pt3,'FontSize',8,...
    'color', [0.5 0.5 0.0], 'HorizontalAlignment','left', 'VerticalAlignment','top')
ylabel('1\sigma Spreading (dB)');
xlabel('Window Size - 2L (\lambda)');
legend('1\sigma spread', 'min(1\sigma spread) - Long-term Fading', 'Inflexion Point of 1\sigma spread - Short-term Fading', 'min(RMSE) - Long-term / Short-term Fading');
title("1\sigma Spread - Long-term Fading")
```

```matlab
%% Plot 1-sigma Spread - Short-term Fading
figure;
plot((vtW-1)*dMod/lambda,sigmaSpreadShort); hold all;
plot((vtW(minIndxSigmaLong)-1)*dMod/lambda, sigmaSpreadShort(minIndxSigmaLong), 'o', 'linewidth', 1, 'Color',[0.9 0.1 0.0])
plot((vtW(infIndxSigmaShort)-1)*dMod/lambda, sigmaSpreadShort(infIndxSigmaShort), 'o', 'linewidth', 1, 'Color',[0.9 0.1 0.8])
plot((vtW(minIndxShort)-1)*dMod/lambda, sigmaSpreadShort(minIndxShort), 'o', 'linewidth', 1, 'Color',[0.5 0.5 0.0])
hold off
pt1 = sprintfc('(%1.2f, %1.2f)', [(vtW(infIndxSigmaShort)-1)*dMod/lambda; sigmaSpreadShort(infIndxSigmaShort)].');
pt2 = sprintfc('(%1.2f, %1.2f)', [(vtW(minIndxSigmaLong)-1)*dMod/lambda; sigmaSpreadShort(minIndxLong)].');
pt3 = sprintfc('(%1.2f, %1.2f)', [(vtW(minIndxShort)-1)*dMod/lambda; sigmaSpreadShort(minIndxShort)].');
text((vtW(infIndxSigmaShort)-1)*dMod/lambda, sigmaSpreadShort(infIndxSigmaShort), pt1,'FontSize',8,...
    'color', [0.9 0.1 0.8], 'HorizontalAlignment','left', 'VerticalAlignment','top')
text((vtW(minIndxSigmaLong)-1)*dMod/lambda, sigmaSpreadShort(minIndxSigmaLong), pt2,'FontSize',8,...
    'color', [0.9 0.1 0.0], 'HorizontalAlignment','right', 'VerticalAlignment','bottom')
text((vtW(minIndxShort)-1)*dMod/lambda, sigmaSpreadShort(minIndxShort), pt3,'FontSize',8,...
    'color', [0.5 0.5 0.0], 'HorizontalAlignment','left', 'VerticalAlignment','top')
ylabel('1\sigma Spreading (dB)');
xlabel('Window Size - 2L (\lambda)');
legend('1\sigma spread', 'min(1\sigma spread) - Long-term Fading', 'Inflexion Point of 1\sigma spread - Short-term Fading', ...
'min(RMSE) - Long-term / Short-term Fading', 'Location','southeast');
title("1\sigma Spread - Short-term Fading")
```

```matlab
% O tamanho da melhor janela é = (quantidade de amostras-1)*espaçamento entre amostras
disp(strcat("O tamanho da janela Modelo (menor RMSE para Long-term/Short-term Fading)   é: ", num2str((vtW(minIndxShort)-1)*dMod/lambda)," ", char(955)))
disp(strcat("O melhor tamanho da janela para 1-sigma Spreading (Long-term Fading) é: ", num2str((vtW(minIndxSigmaLong)-1)*dMod/lambda)," ", char(955)))
disp(strcat("O melhor tamanho da janela para 1-sigma Spreading (Short-term Fading) é: ", num2str((vtW(infIndxSigmaShort)-1)*dMod/lambda)," ", char(955)))
```

- Os pontos de melhor janela para Long-term, Short-term Fading e modelo mostram-se próximos entre si;
- O tamanho de melhor janela pela curva $1\sigma$-Spread (Short-term Fading) apresenta-se mais próximo ao previsto pelo modelo.
