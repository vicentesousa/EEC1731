# Instalação do suporte ao Matlab no Jupyter Notebook (Kernel do Matlab)¶

**Passo 1:** Precisamos da versão do Pyhton3, correta, compatível com o Matlab 2018a. Podemos consultar essa versão pelo link, [Version_Matlab_Version_Python3](https://www.mathworks.com/content/dam/mathworks/mathworks-dot-com/support/sysreq/files/python-compatibility.pdf).  
De acordo com a tabela do link, para o **Matlab 2018a** devemos utilizar a versão **3.5 ou 3.6** do **Python3**.
Então, assumindo que você  instalou o Anaconda mais recente,**Python versão 3.9**,  criaremos o um ambiente virtual chamado **py35**, no qual tem uma instalação do **Python 3.5**.

Abra o terminal,(**ctrl+alt+T**), e execute o comando abaixo:



```matlab
conda create --name py35 python=3.5
```

Em seguida, ative o ambiente:

```matlab
conda activate py35
```

Para confirmar que o ambiente foi ativo, digite python --version. O resultado deve ser semelhante a tela abaixo:


**Passo 2:** Agora vamos instalar o kernel do matlab.

```matlab
sudo pip install matlab_kernel
```

**Passo 3:** Agora, vamos instalar o Matlab Engine para Python, digite (estamos supondo que o seu Matlab está instalado na pasta “/usr/local/MATLAB/R2018a/")

```matlab
cd /usr/local/MATLAB/R2018a/extern/engines/python/
```

Dentro do diretório acima, execute:

```matlab
sudo python3 setup.py install --user
```

Caso apareça uma mensagem de permissão negada ao executar a linha acima, digite o seguinte comando e tente novamente.

```matlab
sudo chmod 777 setup.py
```

**Passo 4:** Mudar as permissões de pasta para o kernel do jupyter notebook funcionar. Digite (os comandos abaixo considerando o **usuário de sua máquina**):

sudo chown **seu_usuario** -R ~/.matlab

sudo chown **seu_usuario** -R ~/.local/lib/
